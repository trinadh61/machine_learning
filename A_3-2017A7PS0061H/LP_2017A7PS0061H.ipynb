```python
import pandas as pd
import numpy as np
from scipy.stats import norm
from matplotlib import pyplot as plt
import math
import random
%matplotlib inline
```

```python
data = pd.read_csv('./dataset_LP_1.csv')
col = []
for i in range(1,len(data.columns)):
    col.append('x' + str(i))
col.append('y')
data.columns = col 
```

```python
test = data.sample(frac = 0.3)
train  = data.drop(test.index)
# print(test, train)
```

```python
#a random

w = [0]*data.shape[1]
w
```

```python
def update_vector(w, row,t, rate = 0.2):
    w[0] = w[0] + (t * rate * 1)
    for i in range(len(row)-1):
        w[i+1] =  w[i+1] + (t * rate * row[i])
    return w

enonch = 0
while enonch < 1000000:
    old_w = list(w)
    miscal = []
    for index, row in train.iterrows():
        enonch = enonch + 1
        sum = w[0]
        for i in range(len(row)-1):
            sum += w[i+1] * row[i]
        if sum >= 0 and row[-1] == 0:
            w = update_vector(w, row, -1)
            miscal.append(row)
        elif sum < 0 and row[-1] == 1:
            w = update_vector(w, row, 1)
            miscal.append(row)
    print(len(miscal))
    if old_w == w:
        break

print(w)
```
