```python
import pandas as pd
import numpy as np
from scipy.stats import norm
from matplotlib import pyplot as plt
import math
%matplotlib inline
```

```python
data = pd.read_csv('./dataset_FLD.csv')
# print((data))
```

```python
col = []
for i in range(1,len(data.columns)):
    col.append('x' + str(i))
col.append('y')
data.columns = col 
```

```python
def get_vector_w(data):
    data0 = data.loc[data['y']==0]
    data1 = data.loc[data['y']==1]
#     print(len(data0) + len(data1))
    
    #building matrices m1 and m2
    
    m0 = []
    m1 = []
    for i in data.columns:
        if i == 'y':
            continue
        else:
            m0.append(np.mean(data0[i]))
            m1.append(np.mean(data1[i]))
    
    m = np.asarray(m0) - np.asarray(m1)
    features = data.drop(columns = ['y'])
    features_cov = np.cov(features, rowvar = False)
    feaures_cov_inv = np.linalg.inv(features_cov)

    return np.dot(feaures_cov_inv, m.T)

w = (get_vector_w(data))
print( ' w = ', w)
# [ 0.01206137  0.0334413  -1.8218901 ]
```

```python
def linear_discriminant_value(data0, data1):
    m1 = np.mean(data0)
    m2 = np.mean(data1)
    std1 = np.std(data0)
    std2 = np.std(data1)
    a = 1/(2* std1 ** 2) - 1/(2 *std2 ** 2)
    b = -m1/(std1 ** 2) + m2/(2 *std2 ** 2)
    c = m1 ** 2/(2* std1 ** 2) - m2 ** 2 /(2 *std2 ** 2) - np.log(std2/std1)
    roots = np.roots([a,b,c])
    for i in roots:
        if m1 < m2 and i > m1 and i < m2 or m1 > m2 and i < m1 and i > m2:
            return i
```

```python
def check_accuracy(data, w, linear_discrim):
    count = 0
    for index, row in data.iterrows():
        sum = -linear_discrim
        for i in range(len(row)-1):
            sum += w[i] * row[i]
        if sum >= 0 and row[-1] == 0 or sum < 0 and row[-1] == 1:
            count += 1
    print(count/data.shape[0]*100)
```

```python
collapse0=[]
data0 = data.loc[data['y']==0]
data1 = data.loc[data['y']==1]
plt.figure(figsize=(10,10))
for i in range(data0.shape[0]):
    collapse0.append(np.dot(w,data0.iloc[i,:-1]))
    
collapse1=[]
for i in range(data1.shape[0]):
    collapse1.append(np.dot(w,data1.iloc[i,:-1]))

```

```python
#ploting the projection of points

plt.scatter(collapse0,[0]*data0.shape[0],color='red', label = 'class 0')
plt.scatter(collapse1,[0]*data1.shape[0],color='green', label = 'class 1')
plt.title('Projection of data onto w vector')
plt.legend()
plt.savefig('fld_1d.jpeg')
plt.clf()
```

```python
#ploting the normal distribution curve

axis_points = list(np.arange(-4,4, 0.05))
plt.figure(figsize=(10,10))
plt.plot(axis_points, norm.pdf(axis_points,np.mean(collapse0),np.std(collapse0) ), color = 'red', label = 'class 0')
plt.plot(axis_points, norm.pdf(axis_points,np.mean(collapse1),np.std(collapse1) ), color = 'green', label = 'class 1')
linear_discriminant = linear_discriminant_value(collapse0, collapse1)

plt.plot([linear_discriminant]*len(axis_points), axis_points, color = 'blue', label = 'Linear Discriminant' )
plt.legend()
plt.savefig('fld_1d_normal_distribution.jpeg')
```

```python
#plotting the data in 3 dimensional space


fig = plt.figure(figsize=(10,10)) 
ax = plt.axes(projection ='3d')
ax.scatter3D(data1['x1'],data1['x2'], data1['x3'], color = 'green', label = 'positive points')
ax.scatter3D(data0['x1'],data0['x2'], data0['x3'], color = 'red', label = 'negative points')
(x, y) = np.meshgrid(np.linspace(-10, 10, 21), np.linspace(-10, 10, 21))
z = (w[0] * x + w[1] * y + linear_discriminant) / w[2]
ax.plot_surface(x,y,z,alpha = 0.5, color = 'black', label = 'Discriminant Hyper Plane')
ax.view_init(30,0)
fig.savefig('flb3d_view1.jpeg')
ax.view_init(0,-30)
fig.savefig('flb3d_view2.jpeg')
```

```python
#checking the accuracy of model

check_accuracy(data, w, linear_discriminant)
```
